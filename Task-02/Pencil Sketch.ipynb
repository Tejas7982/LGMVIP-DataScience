```python
import cv2
import matplotlib.pyplot as plt
```

```python
#reading the image
img = cv2.imread('Car.jpg',0)
```

```python
#conversion to grayscale image
plt.imshow(img, cmap = 'gray')
```

```python
inv = 255-img
```

```python
#conversion to inverted grayscale image
plt.imshow(inv, cmap = 'gray')
```

```python
blur = cv2.GaussianBlur(inv, ksize= (21,21), sigmaX=0, sigmaY=0)
```

```python
#blurred grayscale image
plt.imshow(blur, cmap = 'gray')
```

```python
dodge = lambda image, mask: cv2.divide(image, 255-mask, scale=256)
```

```python
blended = dodge(img, blur)
```

```python
#final pencil sketch image
plt.imshow(blended, cmap= 'gray')
```

```python
cv2.imwrite('Pencil Sketch.jpg',blended)
```
